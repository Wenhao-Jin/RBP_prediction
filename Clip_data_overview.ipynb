```python
import pandas as pd
import numpy as np
import sklearn.cluster
import mpld3
```

```python
raw_table=pd.read_table("public_clip_distributions.csv",sep=',')
raw_table
```

```python
clipseq_table=raw_table
clipseq_table.index=np.array(clipseq_table.iloc[:,0])
clipseq_table=clipseq_table.drop(clipseq_table.columns[0],axis=1)
clipseq_table
```

```python
clipseq_table2=clipseq_table.drop("uncatagorized",axis=0)
clipseq_table2
```

```python
clipseq_table2.loc['all']=0
for i in range(1,clipseq_table2.shape[0]):
    clipseq_table2.loc['all']+=clipseq_table2.iloc[i]
clipseq_table2
```

```python
clipseq_norm=clipseq_table2
clipseq_norm.iloc[1:,:]=clipseq_norm.iloc[1:,:]/clipseq_norm.iloc[0,:]
clipseq_norm=clipseq_norm.drop('all',axis=0)
```

```python
clipseq_norm
```

```python
import matplotlib.pyplot as plt
```

```python
clipseq_norm.plot()
plt.show()
```

```python
clipseq_matrix=np.array(clipseq_norm.transpose())
```

```python
hierarchical_clustering=sklearn.cluster.AgglomerativeClustering(linkage="average")
```

```python
hierarchical_clustering.fit(clipseq_matrix)
```

```python
#mpld3.enable_notebook()
from scipy.cluster import hierarchy
import pylab

label_dict={0:'CDS', 1:'distintron500', 2:"5' UTR", 3:'proxintron500', 4:"3' UTR"}
fig = pylab.figure(figsize=(12,8))

#plot the first clustering (cluster the RBP)
ax1 = fig.add_axes([0.3,0.71,0.6,0.2])
Linkage_matrix = hierarchy.linkage(clipseq_matrix, method='centroid')
Tree1 = hierarchy.dendrogram(Linkage_matrix, orientation='top')
ax1.set_xticks([])
#ax1.set_yticks([])

#plot the second clustering (cluster the region labels)
ax2 = fig.add_axes([0.09,0.1,0.1,0.6])
Linkage_matrix = hierarchy.linkage(clipseq_matrix.transpose(), method='centroid')
Tree2 = hierarchy.dendrogram(Linkage_matrix, orientation='right')
#ax2.set_xticks([])
ax2.set_yticks([])

#plot the heatmap
axmatrix = fig.add_axes([0.3,0.1,0.6,0.6])
idx1 = Tree1['leaves']
idx2 = Tree2['leaves']
#print(idx2)
temp_matrx = clipseq_matrix.transpose()
temp_matrx = temp_matrx[:,idx1]
temp_matrx = temp_matrx[idx2,:]
im = axmatrix.matshow(temp_matrx, aspect='auto', origin='lower', cmap=pylab.cm.YlGnBu)
axmatrix.set_xticks([])
#axmatrix.set_yticks([])
#axmatrix.yaxis.set_label([label_dict[idx2[0]],label_dict[idx2[1]],label_dict[idx2[2]],label_dict[idx2[3]],label_dict[idx2[4]]])
axmatrix.set_yticks(range(5))
axmatrix.set_yticklabels([label_dict[idx2[0]],label_dict[idx2[1]],label_dict[idx2[2]],label_dict[idx2[3]],label_dict[idx2[4]]], minor=False)
axmatrix.yaxis.set_label_position('left')
axmatrix.yaxis.tick_left()

#plot the color bar
axcolor = fig.add_axes([0.91,0.1,0.02,0.6])
pylab.colorbar(im, cax=axcolor)
plt.savefig('clustering_RBPvsBindRegion.png')
from IPython.display import Image
i=Image(filename='clustering_RBPvsBindRegion.png')
i
#mpld3.display()
```
