```python
import pandas as pd
import numpy as np
import re
```

```python
public_clip_table=pd.read_csv('public_clip_distributions.csv')
public_clip_table2=pd.read_csv('public_clip_distributions.csv')
public_clip_table.index=np.array(public_clip_table.loc[:,public_clip_table.columns[0]])
public_clip_table=public_clip_table.drop(public_clip_table.columns[0],axis=1)
public_clip_table2.index=np.array(public_clip_table2.loc[:,public_clip_table2.columns[0]])
public_clip_table2=public_clip_table2.drop(public_clip_table2.columns[0],axis=1)
```

```python
#public_clip_table
```

```python
index_table=pd.read_table('Public CLIP _Database_ - Sheet1.tsv')
index_table2=index_table[['SRA','replicate','gencode_v17_id','Species','hg19_equivalent_gencode','Modification']]
```

```python
#correct the items with SRR* to sample name
columns=[]
p=re.compile("^SRR.*")
for i in range(0, len(public_clip_table.columns)):
    if(p.search(public_clip_table.columns[i])):
        tmp=index_table2.ix[:,'replicate'][index_table2['SRA']==public_clip_table.columns[i]].get_values()
        if(tmp):
            columns.append((index_table2.ix[:,'replicate'][index_table2['SRA']==public_clip_table.columns[i]].get_values())[0])
        else:
            columns.append(np.nan)
#        columns.append((index_table2.ix[:,'replicate'][index_table2['SRA']==public_clip_table.columns[i]].get_values())[0])
    else:
        columns.append(public_clip_table.columns[i])
len(columns)
public_clip_table.columns=columns
#public_clip_table
```

```python
#find ensembl id for each sample
ens_columns=[]
for sample in public_clip_table2.columns:
    tmp = index_table2['gencode_v17_id'][index_table2['replicate']==sample].get_values()
    if(len(tmp)):
        ens_columns.append(np.array(tmp)[0])
    else:
        tmp2=index_table2['gencode_v17_id'][index_table2['SRA']==sample].get_values()
        if(len(tmp2)):
            ens_columns.append(np.array(tmp2)[0])
        else:
            print sample
            ens_columns.append(np.nan)
ens_hg19_columns=[]
for sample in public_clip_table2.columns:
    tmp3 = index_table2['hg19_equivalent_gencode'][index_table2['replicate']==sample].get_values()
    if(len(tmp3)):
        ens_hg19_columns.append(np.array(tmp3)[0])
    else:
        tmp4=index_table2['hg19_equivalent_gencode'][index_table2['SRA']==sample].get_values()
        if(len(tmp4)):
            ens_hg19_columns.append(np.array(tmp4)[0])
        else:
            print sample
            ens_hg19_columns.append(np.nan)
Modification_columns=[]
for sample in public_clip_table2.columns:
    tmp5 = index_table2['Modification'][index_table2['replicate']==sample].get_values()
    if(len(tmp5)):
        Modification_columns.append(np.array(tmp5)[0])
    else:
        tmp6=index_table2['Modification'][index_table2['SRA']==sample].get_values()
        if(len(tmp6)):
            Modification_columns.append(np.array(tmp6)[0])
        else:
#            print sample
            Modification_columns.append(np.nan)
#len(Modification_columns)
```

```python
public_clip_table3=public_clip_table2.transpose()
#public_clip_table2.loc['ensembl_id']=ens_columns
#public_clip_table2.loc['ens_hg19_id']=ens_hg19_columns
public_clip_table3['ensembl_id']=ens_columns
public_clip_table3['ensembl_hg19_id']=ens_hg19_columns
public_clip_table3['Modification']=Modification_columns
#public_clip_table3.to_csv('public_clip_distribution_revised2.csv')
public_clip_table3
```
